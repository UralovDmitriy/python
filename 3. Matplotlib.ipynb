````python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
````

````python
plt.plot(X, C, color="blue", linewidth=1)
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--")
plt.show()
````

````python
plt.plot(X, C, color="blue", linewidth=1, label='cos')
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--", label='sin')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1])

plt.legend()
plt.show()
````

````python
import numpy as np
a = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt")

import matplotlib.pyplot as plt

plt.plot(a[:, 0], a[:, 1], label = "Hares")
plt.plot(a[:, 0], a[:, 2], label = "Lynxes")
plt.plot(a[:, 0], a[:, 3], label = "Carrots")
print('Markers show the top 2 years for each species when they had the lowest populations')
plt.xticks(np.arange(a[:, 0].min(), a[:, 0].max(), 3))
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
pos1 = np.argmin(a[:, 1], axis = 0)
pos2 = np.argmin(a[:, 2], axis = 0)
pos3 = np.argmin(a[:, 3], axis = 0)
plt.scatter(a[pos1, 0], a[:, 1].min(), s=70, c='b')
plt.scatter(a[pos2, 0], a[:, 2].min(), s=70, c='orange')
plt.scatter(a[pos3, 0], a[:, 3].min(), s=70, c='g')
d1 = np.delete(a[:, 1], pos1)
d2 = np.delete(a[:, 2], pos2)
d3 = np.delete(a[:, 3], pos3)
b1 = np.argmin(d1, axis = 0)
b2 = np.argmin(d2, axis = 0)
b3 = np.argmin(d3, axis = 0)
if (b1 >= pos1):
    b1 += 1
if (b2 >= pos2):
    b2 += 1
if (b3 >= pos3):
    b3 += 1
plt.scatter(a[b1, 0], d1.min(), s=70, c='b')
plt.scatter(a[b2, 0], d2.min(), s=70, c='orange')
plt.scatter(a[b3, 0], d3.min(), s=70, c='g')
plt.show()
````

````python
import numpy as np
a = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt")

#экстраполяция с помощью многочлена Лагранжа
n = 20
x = 1920
l = np.empty(n)
for i in range(n):
    b = np.delete(a[0:-1, 0], i)
    l[i] = (x - b).prod() / (a[i, 0] - b).prod()
ans = (a[0:-1, 1] * l).sum()
print(ans)
print('Error = ' + str(np.fabs(ans - a[-1, 1]) / a[-1, 1] * 100) + ' %')

#экстраполяция с помощью квадратичной функции, построенной по трём точкам
a2 = (a[-2, 1] - a[-4, 1]) / ((a[-2, 0] - a[-4, 0])*(a[-2, 0] - a[-3, 0])) 
a2 -= (a[-3, 1] - a[-4, 1]) / ((a[-3, 0] - a[-4, 0])*(a[-2, 0] - a[-3, 0]))
a1 = (a[-3, 1] - a[-4, 1]) / (a[-3, 0] - a[-4, 0]) - a2 * (a[-3, 0] + a[-4, 0])
a0 = a[-4, 1] - a1 * a[-4, 0] - a2 * a[-4, 0] * a[-4, 0]
ans = a0 + a1 * x + a2 * x * x
print(ans)
print('Error = ' + str(np.fabs(ans - a[-1, 1]) / a[-1, 1] * 100) + ' %')
````
